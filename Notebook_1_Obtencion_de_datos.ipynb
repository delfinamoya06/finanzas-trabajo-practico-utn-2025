```python
# Descarga y armado del dataset
# Descargamos los datos para las acciones de bebidas y el S&P 500
TICKERS_MAIN = ["KO", "PEP", "KDP", "MNST", "^GSPC"]
CSV_OUT = f"beverage_stocks_{date.today().strftime('%Y-%m-%d')}.csv"
START_DATE = "2020-07-01"

data = yf.download(TICKERS_MAIN, start=START_DATE, interval="1d",
                   auto_adjust=False, progress=False, threads=True)
adj = data["Adj Close"].ffill()
adj = adj.rename(columns={'^GSPC': 'SPY'})

# Correct the column selection to use 'SPY' instead of '^GSPC'
prices_df = adj[["KO", "PEP", "KDP", "MNST", "SPY"]].copy()
prices_df.to_csv(CSV_OUT, index_label="Date")
print(f"‚úÖ CSV generado ‚Üí {CSV_OUT}")
print("Columnas:", prices_df.columns.tolist())

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 3) Funciones de riesgo y correlaciones ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
def daily_returns(price_df: pd.DataFrame) -> pd.DataFrame:
    """Calcula los retornos logar√≠tmicos diarios."""
    return np.log(price_df / price_df.shift(1)).dropna()

def var_cvar(ret: pd.Series, alpha=0.95):
    """Calcula el Value at Risk (VaR) y el Conditional VaR (CVaR)."""
    q = np.quantile(ret, 1 - alpha)
    return q, ret[ret <= q].mean()

def max_drawdown(price: pd.Series):
    """Calcula el Max Drawdown, su pico y valle."""
    roll_max = price.cummax()
    dd = price / roll_max - 1
    return dd.min(), price.loc[:dd.idxmin()].idxmax(), dd.idxmin()

def corr_heatmap(ret_df: pd.DataFrame, title="Matriz de correlaciones"):
    """Genera un heatmap de correlaci√≥n."""
    corr = ret_df.corr()
    fig, ax = plt.subplots(figsize=(8, 6))
    im = ax.imshow(corr, cmap="RdBu", vmin=-1, vmax=1)
    ax.set_xticks(range(len(corr))); ax.set_xticklabels(corr.columns, rotation=90)
    ax.set_yticks(range(len(corr))); ax.set_yticklabels(corr.index)
    plt.colorbar(im, ax=ax); ax.set_title(title); plt.tight_layout(); plt.show()

# Efficient Frontier v-3.3  (psd_wrap + anti-IndexError)
def efficient_frontier(ret_df, rf=0.0,
                       min_w=0.05, max_w=0.40,
                       trans_cost=0.005, jitter=1e-8):
    """Calcula y grafica la frontera eficiente de Markowitz."""
    valid = ret_df.var()[lambda v: v>0].index
    ret_df = ret_df[valid]
    if ret_df.shape[1] < 2:
        raise ValueError("‚â•2 activos con datos v√°lidos requeridos")
    mu = (ret_df.mean()*252).to_numpy()
    Sigma = np.atleast_2d(ret_df.cov()*252)+np.eye(len(mu))*jitter
    Sigma = cp.psd_wrap(Sigma)

    n, w = len(mu), cp.Variable(len(mu))
    gamma = cp.Parameter(nonneg=True)
    risk = cp.quad_form(w, Sigma)
    ret = mu @ w - trans_cost
    cons = [cp.sum(w)==1, w>=min_w, w<=max_w]

    curve, solvers = [], [cp.OSQP, cp.ECOS, cp.SCS]
    for g in np.logspace(-2,3,60):
        gamma.value = g
        solved = False
        for s in solvers:
            try:
                cp.Problem(cp.Maximize(ret-gamma*risk),cons).solve(solver=s,verbose=False)
                solved=True; break
            except cp.SolverError: pass
        if solved and w.value is not None:
            curve.append([ret.value, np.sqrt(risk.value), *w.value])

    ef = pd.DataFrame(curve, columns=["Return","Vol"]+list(ret_df.columns))
    ef["Sharpe"] = (ef["Return"]-rf)/ef["Vol"]
    return ef, ef.loc[ef["Sharpe"].idxmax()], ef.loc[ef["Vol"].idxmin()]


# Indicadores t√©cnicos & backtesting
def add_technical_indicators(df: pd.DataFrame, ticker: str):
    """A√±ade indicadores t√©cnicos a un DataFrame de precios."""
    px = df[ticker]
    ind = pd.DataFrame(index=px.index)
    ind["Close"] = px
    ind["SMA50"]  = ta.trend.sma_indicator(px, 50)
    ind["SMA200"] = ta.trend.sma_indicator(px, 200)
    macd = ta.trend.MACD(px); ind["MACD"] = macd.macd(); ind["MACD_signal"] = macd.macd_signal()
    ind["RSI14"] = ta.momentum.rsi(px, 14)
    bb = ta.volatility.BollingerBands(px, 20, 2)
    ind["BB_high"] = bb.bollinger_hband(); ind["BB_low"] = bb.bollinger_lband()
    return ind

def backtest_ma_cross(df_ind: pd.DataFrame, start_cash=1_000):
    """Realiza un backtesting de la estrategia de cruce de medias m√≥viles."""
    pos,cash,shares = 0,start_cash,0; equity=[]
    for _, r in df_ind.iterrows():
        p = r["Close"]
        if np.isnan(r["SMA50"]) or np.isnan(r["SMA200"]):
            equity.append(cash+shares*p); continue
        if r["SMA50"]>r["SMA200"] and pos==0: shares,cash,pos=cash/p,0,1
        elif r["SMA50"]<r["SMA200"] and pos==1: cash,shares,pos=shares*p,0,0
        equity.append(cash+shares*p)
    equity = pd.Series(equity, index=df_ind.index, name="Equity")
    ret = equity.pct_change().dropna()
    sharpe = np.sqrt(252)*ret.mean()/ret.std() if ret.std()>0 else np.nan
    dd,_,_ = max_drawdown(equity)
    return equity, sharpe, dd
```

```python
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ simulaci√≥n de la frontera eficiente ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
def efficient_frontier_sim(ret_df, rf=0.0,
                           min_w=0.05, max_w=0.40,
                           n_port=20_000, trans_cost=0.005):
    """
    Calcula la frontera eficiente y portafolios √≥ptimos mediante
    una simulaci√≥n de Montecarlo.

    El objetivo es encontrar portafolios que cumplan con las
    siguientes restricciones:
      - 5 % <= peso de cada activo (w) <= 40 %
      - La suma de los pesos es igual a 1 (sum(w) = 1)

    Par√°metros:
    -----------
    ret_df (DataFrame): DataFrame con los retornos de los activos.
    rf (float): Tasa libre de riesgo, por defecto 0.0.
    min_w (float): Peso m√≠nimo de un activo en el portafolio.
    max_w (float): Peso m√°ximo de un activo en el portafolio.
    n_port (int): N√∫mero de portafolios a simular.
    trans_cost (float): Costo de transacci√≥n.

    Devuelve:
    ---------
    frontier (DataFrame): Portafolios que forman la frontera eficiente.
                          Ordenados por volatilidad de menor a mayor.
    p_max (Series): Portafolio con el m√°ximo ratio de Sharpe.
    p_min (Series): Portafolio con la m√≠nima volatilidad.
    """
    # 1. Preparar los datos
    # Aplanar las columnas si son multi-√≠ndice y obtener los nombres de los activos
    ret_df = ret_df.copy()
    ret_df.columns = [
        col[-1] if isinstance(col, tuple) else str(col)
        for col in ret_df.columns
    ]
    asset_names = list(ret_df.columns)

    # 2. Calcular estad√≠sticas anuales
    # Se anualizan la media de los retornos y la matriz de covarianza
    mu = ret_df.mean() * 252
    cov = ret_df.cov() * 252

    # 3. Realizar la simulaci√≥n de portafolios
    sim_results = []
    rng = np.random.default_rng()

    for _ in range(n_port):
        # Generar pesos aleatorios y asegurarse de que sumen 1
        weights = rng.uniform(min_w, max_w, len(asset_names))
        weights /= weights.sum()

        # Calcular retorno y volatilidad del portafolio
        port_return = np.dot(weights, mu) - trans_cost
        port_vol = np.sqrt(np.dot(weights, cov @ weights))

        # Calcular ratio de Sharpe
        sharpe_ratio = (port_return - rf) / port_vol if port_vol > 0 else np.nan

        # Guardar los resultados
        sim_results.append([port_return, port_vol, sharpe_ratio, *weights])

    # Convertir resultados a un DataFrame
    sims = pd.DataFrame(sim_results, columns=["Return", "Vol", "Sharpe"] + asset_names)

    # 4. Encontrar los portafolios √≥ptimos
    # Obtener la frontera eficiente (envolvente superior)
    frontier = sims.loc[sims.groupby("Vol")["Return"].idxmax()].reset_index(drop=True)

    # Encontrar el portafolio de m√°ximo Sharpe y el de m√≠nima volatilidad
    p_max = sims.loc[sims["Sharpe"].idxmax()]
    p_min = sims.loc[sims["Vol"].idxmin()]

    return frontier, p_max, p_min
```

```python
# 0 ¬∑ CONFIGURACI√ìN Y LIBRER√çAS¬†‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import date
from scipy.stats import jarque_bera

# Configuraci√≥n de visualizaci√≥n
sns.set_style("whitegrid")
plt.rcParams["figure.dpi"] = 110

# Tickers de las empresas a analizar y el √≠ndice de referencia (S&P 500 ETF)
TICKERS_MAIN = ["KO", "PEP", "KDP", "MNST", "SPY"]
START_DATE = "2020-07-01"
CSV_OUT = f"beverage_stocks_{date.today().strftime('%Y-%m-%d')}.csv"

# --- FUNCIONES DE AN√ÅLISIS ----------------------------------------------------
def var_cvar(ret: pd.Series, alpha=0.95):
    """Calcula el Value at Risk (VaR) y el Conditional VaR (CVaR)."""
    q = np.quantile(ret, 1 - alpha)
    return q, ret[ret <= q].mean()

def max_drawdown(price: pd.Series):
    """Calcula el Max Drawdown, su pico y valle."""
    roll_max = price.cummax()
    dd = price / roll_max - 1
    return dd.min(), price.loc[:dd.idxmin()].idxmax(), dd.idxmin()

def corr_heatmap(ret_df: pd.DataFrame, title="Matriz de correlaciones"):
    """Genera un heatmap de correlaci√≥n."""
    corr = ret_df.corr()
    fig, ax = plt.subplots(figsize=(8, 6))
    im = ax.imshow(corr, cmap="RdBu", vmin=-1, vmax=1)
    ax.set_xticks(range(len(corr))); ax.set_xticklabels(corr.columns, rotation=90)
    ax.set_yticks(range(len(corr))); ax.set_yticklabels(corr.index)
    plt.colorbar(im, ax=ax); ax.set_title(title); plt.tight_layout(); plt.show()

# --- DESCARGA Y PREPARACI√ìN DE DATOS ------------------------------------------
print("1) Descargando datos de yfinance...")
data = yf.download(TICKERS_MAIN, start=START_DATE, interval="1d",
                   auto_adjust=False, progress=False, threads=True)
adj = data["Adj Close"].ffill()

prices_df = adj[TICKERS_MAIN].copy()
prices_df.to_csv(CSV_OUT, index_label="Date")
print(f"‚úÖ CSV generado ‚Üí {CSV_OUT}")
print("Columnas:", prices_df.columns.tolist())

# Calcular retornos diarios logar√≠tmicos
ret_df = np.log(prices_df / prices_df.shift(1)).dropna()

# --- C√ÅLCULO DE M√âTRICAS DE RIESGO --------------------------------------------
print("\n2) Calculando m√©tricas de riesgo...")
risk_metrics_list = []
for ticker in TICKERS_MAIN:
    # Calcular VaR, CVaR, Max Drawdown y la prueba de Jarque-Bera
    var95, cvar95 = var_cvar(ret_df[ticker])
    max_dd, _, _ = max_drawdown(prices_df[ticker])
    jb_statistic, jb_p_value = jarque_bera(ret_df[ticker])

    # A√±adir los resultados a la lista
    risk_metrics_list.append([
        ticker,
        var95,
        cvar95,
        max_dd,
        jb_p_value
    ])

# Crear y mostrar el DataFrame con los resultados
risk_analysis_df = pd.DataFrame(
    risk_metrics_list,
    columns=["Ticker", "VaR 95%", "CVaR 95%", "Max Drawdown", "JB p-value"]
)
print("‚úÖ An√°lisis de riesgo completado:")
display(risk_analysis_df)

# --- AN√ÅLISIS DE CORRELACIONES ------------------------------------------------
print("\n3) Visualizando matriz de correlaciones...")
corr_heatmap(ret_df, title="Matriz de Correlaciones de Retornos")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import jarque_bera

# ----------------- PAR√ÅMETROS -----------------
TICKERS_MAIN = ["KO", "PEP", "KDP", "MNST", "SPY"]
START_DATE = "2020-01-01"
END_DATE = "2024-12-31"

# ----------------- FUNCIONES AUXILIARES -----------------
def get_historical_data(tickers, start, end):
    """Descarga datos hist√≥ricos de yfinance y devuelve los precios de cierre ajustados."""
    try:
        prices = yf.download(tickers, start=start, end=end, auto_adjust=True)['Close']
        if prices.empty:
            raise ValueError("No se pudieron descargar los datos para los tickers especificados.")
        return prices
    except Exception as e:
        print(f"Error descargando datos: {e}")
        return None

def daily_returns(prices):
    """Calcula los retornos diarios logar√≠tmicos."""
    return np.log(prices / prices.shift(1)).dropna()

def plot_normalized_prices(prices_df, title):
    """Grafica los precios normalizados a base 100."""
    plt.figure(figsize=(12, 6))
    normalized_prices = (prices_df / prices_df.iloc[0]) * 100
    normalized_prices.plot(ax=plt.gca())
    plt.title(title, fontsize=16)
    plt.xlabel("Fecha")
    plt.ylabel("Precio Normalizado (Base 100)")
    plt.legend(normalized_prices.columns)
    plt.grid(True)
    plt.show()

def calculate_drawdown(returns):
    """Calcula el drawdown de una serie de retornos."""
    cumulative_returns = (1 + returns).cumprod()
    peak = cumulative_returns.cummax()
    drawdown = (cumulative_returns - peak) / peak
    max_drawdown = drawdown.min()
    return drawdown, max_drawdown

def calculate_risk_metrics(returns):
    """Calcula VaR, CVaR y Drawdown M√°ximo."""
    var_95 = np.percentile(returns, 5) # Percentil 5 para VaR 95%
    cvar_95 = returns[returns < var_95].mean()
    _, max_drawdown = calculate_drawdown(returns)
    return var_95, cvar_95, max_drawdown

# ----------------- L√ìGICA PRINCIPAL -----------------
if __name__ == "__main__":
    # --- Descargar datos y calcular retornos ---
    prices_df = get_historical_data(TICKERS_MAIN, START_DATE, END_DATE)
    if prices_df is None:
        exit()

    rets = daily_returns(prices_df)

    # ----------------------------------------------------------------
    # 1. GR√ÅFICOS DE PRECIOS NORMALIZADOS
    # ----------------------------------------------------------------
    print("\n--- 1. Gr√°ficos de Precios Normalizados (Base 100) ---")
    plot_normalized_prices(prices_df, "Precios Normalizados de Activos")

    # ----------------------------------------------------------------
    # 2. IDENTIFICACI√ìN Y AN√ÅLISIS DE EVENTOS
    # ----------------------------------------------------------------
    print("\n--- 2. Identificaci√≥n y Explicaci√≥n de Eventos ---")
    print("El gr√°fico de precios normalizados permite identificar visualmente "
          "eventos significativos. Aqu√≠ se explican 3 eventos potenciales:")

    # Evento 1: Pandemia de COVID-19 (Marzo 2020)
    print("\nEvento 1: Ca√≠da por la Pandemia de COVID-19 (Marzo 2020)")
    print(" - **Impacto:** Todos los activos, incluido el S&P 500 (SPY), experimentaron una ca√≠da abrupta y sincronizada. "
          "Esto se debi√≥ a la incertidumbre econ√≥mica global, el confinamiento y el cierre de negocios. "
          "Las empresas de bienes de consumo (KO, PEP) mostraron cierta resiliencia, but no fueron inmunes.")

    # Evento 2: Crisis de la cadena de suministro e inflaci√≥n (2022)
    print("\nEvento 2: Crisis de la cadena de suministro e inflaci√≥n (2022)")
    print(" - **Impacto:** En 2022, el mercado en general (SPY) cay√≥ debido a la inflaci√≥n persistente y la subida de tipos de inter√©s de la Fed. "
          "Empresas como KO y PEP, al ser de consumo b√°sico, pudieron transferir parte de los costes a los consumidores, "
          "mostrando una ca√≠da menos severa que el mercado en general.")

    # Evento 3: Boom del mercado tecnol√≥gico y de IA (2023-2024)
    print("\nEvento 3: Boom del mercado tecnol√≥gico y de IA (2023-2024)")
    print(" - **Impacto:** El mercado (SPY) experiment√≥ una fuerte recuperaci√≥n impulsada por el optimismo en torno a la inteligencia artificial. "
          "Los activos de consumo como KDP y MNST, aunque se beneficiaron del crecimiento econ√≥mico, "
          "crecieron a un ritmo m√°s moderado en comparaci√≥n con los l√≠deres tecnol√≥gicos, lo que se refleja en la divergencia del gr√°fico.")

    # ----------------------------------------------------------------
    # 3. AN√ÅLISIS DE ESTACIONALIDAD POR TRIMESTRES
    # ----------------------------------------------------------------
    print("\n--- 3. An√°lisis de Estacionalidad por Trimestres ---")
    rets_quarterly = rets.copy()
    rets_quarterly['quarter'] = rets_quarterly.index.quarter
    rets_quarterly['year'] = rets_quarterly.index.year # Add year column
    # Include 'year' in id_vars to differentiate quarters across different years if needed
    rets_melted = pd.melt(rets_quarterly, id_vars=['quarter', 'year'], value_vars=TICKERS_MAIN, var_name='Ticker', value_name='Return')

    plt.figure(figsize=(12, 6))
    # Use 'Ticker' as hue since that's the column name after melting
    sns.boxplot(x='quarter', y='Return', hue='Ticker', data=rets_melted)
    plt.title("Retornos Diarios por Trimestre del A√±o", fontsize=16)
    plt.xlabel("Trimestre")
    plt.ylabel("Retorno Diario")
    plt.grid(True)
    plt.show()

    # Gr√°fico 2: Heatmap de retornos trimestrales promedio
    quarterly_avg_returns = rets_quarterly.groupby('quarter')[TICKERS_MAIN].mean()
    plt.figure(figsize=(10, 6))
    sns.heatmap(quarterly_avg_returns.T, annot=True, fmt=".4f", cmap="coolwarm", center=0)
    plt.title("Retornos Diarios Promedio por Trimestre y Ticker", fontsize=16)
    plt.xlabel("Trimestre")
    plt.ylabel("Ticker")
    plt.show()

    print("El gr√°fico de cajas muestra la distribuci√≥n de los retornos diarios para cada trimestre. "
          "El heatmap complementa este an√°lisis visualizando los retornos promedio de cada activo en cada trimestre, "
          "facilitando la identificaci√≥n de patrones de estacionalidad.")

    # ----------------------------------------------------------------
    # 4. C√ÅLCULO DE M√âTRICAS DE RIESGO
    # ----------------------------------------------------------------
    print("\n--- 4. M√©tricas de Riesgo ---")
    risk_metrics = pd.DataFrame(index=TICKERS_MAIN, columns=['VaR 95%', 'CVaR 95%', 'Max Drawdown'])
    for ticker in TICKERS_MAIN:
        var, cvar, mdd = calculate_risk_metrics(rets[ticker])
        risk_metrics.loc[ticker] = [var, cvar, mdd]

    print(risk_metrics)
    print("\n**Interpretaci√≥n:**")
    print(" - **VaR 95%:** Es la p√©rdida m√°xima que no se espera superar con una probabilidad del 95% en un d√≠a. "
          "Una VaR de -2% significa que en el 95% de los d√≠as, la p√©rdida no superar√° el 2%.")
    print(" - **CVaR (Expected Shortfall):** Es la p√©rdida promedio en el peor 5% de los escenarios. "
          "Ofrece una medida de riesgo m√°s conservadora que VaR, ya que considera la magnitud de las p√©rdidas extremas.")
    print(" - **Drawdown M√°ximo:** Es la mayor ca√≠da porcentual desde un pico a un m√≠nimo en un per√≠odo de tiempo.")

    # Visualizaci√≥n del Drawdown
    plt.figure(figsize=(12, 6))
    for ticker in TICKERS_MAIN:
        drawdown, _ = calculate_drawdown(rets[ticker])
        drawdown.plot(label=ticker)
    plt.title("Drawdown Hist√≥rico", fontsize=16)
    plt.xlabel("Fecha")
    plt.ylabel("P√©rdida sobre el Pico (%)")
    plt.legend()
    plt.grid(True)
    plt.show()

    # ----------------------------------------------------------------
    # 5. TEST DE NORMALIDAD Y AN√ÅLISIS DE COLAS
    # ----------------------------------------------------------------
    print("\n--- 5. Test de Normalidad (Jarque-Bera) y An√°lisis de Colas ---")
    for ticker in TICKERS_MAIN:
        jb_test = jarque_bera(rets[ticker])
        print(f"\nResultados Jarque-Bera para {ticker}:")
        print(f" - Estad√≠stica de prueba: {jb_test[0]:.2f}")
        print(f" - p-value: {jb_test[1]:.4f}")

        # Hip√≥tesis nula (H0): Los datos provienen de una distribuci√≥n normal.
        if jb_test[1] < 0.05:
            print(" - **Conclusi√≥n:** Se rechaza la hip√≥tesis nula. Los retornos no son normales.")
        else:
            print(" - **Conclusi√≥n:** No se puede rechazar la hip√≥tesis nula. Los retornos son normales.")

        print(f" - **Sesgo (Skewness):** {rets[ticker].skew():.2f}")
        print(f" - **Curtosis (Kurtosis):** {rets[ticker].kurtosis():.2f}")

    print("\nAn√°lisis de colas:")
    print(" - **Skewness (Sesgo):** Una skewness cercana a 0 indica simetr√≠a. Un valor negativo indica una 'cola' m√°s larga "
          "a la izquierda (m√°s p√©rdidas extremas).")
    print(" - **Curtosis (Kurtosis):** La curtosis de una distribuci√≥n normal es 3. Un valor superior a 3 (leptocurtosis) "
          "indica que los datos tienen colas m√°s 'gordas' que una normal. Esto significa que los eventos extremos (tanto positivos como negativos) "
          "son m√°s probables de lo que predecir√≠a un modelo normal.")

    # Gr√°fico de la distribuci√≥n de retornos vs. una distribuci√≥n normal
    for ticker in TICKERS_MAIN:
        plt.figure(figsize=(8, 5))
        sns.histplot(rets[ticker], kde=True, stat='density', label=ticker, bins=50)
        mu, sigma = rets[ticker].mean(), rets[ticker].std()
        x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
        plt.plot(x, (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)/sigma)**2), 'r--', label='Distribuci√≥n Normal')
        plt.title(f"Distribuci√≥n de Retornos Diarios para {ticker}")
        plt.legend()
        plt.show()
```
